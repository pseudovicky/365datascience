```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv("movie.csv")
df = pd.DataFrame(data)
df
```

```python
# Convert 'release_date' to datetime format
df['release_date'] = pd.to_datetime(df['release_date'])

# Extract release year from the date
df['release_year'] = df['release_date'].dt.year
```

```python
# Plot 1: Distribution of Movie Ratings (vote_average)
plt.figure(figsize=(10, 6))
sns.histplot(df['vote_average'], bins=10, kde=True, color='blue')
plt.title('Distribution of Movie Ratings (Vote Average)')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot 2: Popularity vs Vote Count
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['popularity'], y=df['vote_count'], hue=df['vote_average'], palette='coolwarm', size=df['vote_average'], sizes=(50, 200))
plt.title('Popularity vs Vote Count')
plt.xlabel('Popularity')
plt.ylabel('Vote Count')
plt.show()
```

```python
# Plot 3: Movies by Release Year
plt.figure(figsize=(10, 6))
sns.countplot(x='release_year', data=df, palette='viridis')
plt.title('Number of Movies Released by Year')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.xticks(rotation=90) 
plt.show()
```
